What is Linear Regression?

Linear Regression is used when:

Output is continuous (price, marks, salary)

Relationship is roughly linear

Example:

Predict house price based on size (sqft)

Linear regression assumes this relationship:

ùë¶=ùëöùë•+ùëè

In ML terms:

ùë¶^=ùë§ùëã+ùëè

Where:

w ‚Üí weight (slope)
b ‚Üí bias (intercept)
≈∑ ‚Üí predicted value
üëâ Model learns w and b from data.



```python
## simple Linear Regression

import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

w = 0
b = 0
learning_rate = 0.01

for epoch in range(1000):
    y_pred = w * x + b
    error = y_pred - y
    w_grad = (2/len(x)) * np.dot(error, x)
    b_grad = (2/len(x)) * np.sum(error)
    w -= learning_rate * w_grad
    b -= learning_rate * b_grad

print("Trained weights:", w , "bias:", b)    
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([3, 5, 7, 9, 11, 13])

model = LinearRegression()
model.fit(X, y)

print(model.predict([[9]]))
print("Weights:", model.coef_, "Bias:", model.intercept_)

```

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data[:5])
print(iris.target[:5])
```

```python
model = LinearRegression()
model.fit(iris.data, iris.target)
print(model.predict([iris.data[0]]))
```

```python
print("weights:", model.coef_)
print("intercept:", model.intercept_)
```

# Steps in Machine Learning for Prediction  
## Using Simple Linear Regression

---

### Step 1: Import the Required Libraries
We import the necessary Python libraries required for numerical computation and machine learning model creation.

- **NumPy**: Used for handling arrays and numerical operations  
- **scikit-learn**: Provides the Linear Regression model

---

### Step 2: Creating the Dataset
In this step, we define:
- **Input feature (X)** ‚Üí independent variable  
- **Target variable (y)** ‚Üí dependent variable  

The dataset represents the relationship between input and output values.

---

### Step 3: Creating the Model
Here, we create an instance of the **Linear Regression model**.  
At this stage, the model is empty and has not learned anything from the data.

---

### Step 4: Training the Model
The model is trained using the `.fit(X, y)` method.

During training:
- The model learns the best-fit line  
- It calculates optimal values for **weight (slope)** and **bias (intercept)**  
- The error between predicted and actual values is minimized

---

### Step 5: Making a Prediction
After training, the model can predict output values for new, unseen input data using the `.predict()` method.

---

### Step 6: Accessing Learned Parameters
Once the model is trained, we can access:
- **Coefficient (`coef_`)** ‚Üí Represents the slope (weight)  
- **Intercept (`intercept_`)** ‚Üí Represents the bias  

These parameters define the learned linear equation.
