KNN for Classification

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

data = {
    'student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'hours_studied': [2, 4, 1, 5, 3],
    'exam_score': ['failed', 'passed', 'failed', 'passed', 'failed']
}
df = pd.DataFrame(data)
X = df[['hours_studied']]
y = df['exam_score']    
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
hours = input("Enter number of hours studied: ")

predictions = model.predict([[float(hours)]])
print(f"The predicted exam result is: {predictions[0]}")
```

```python
from sklearn.neighbors import KNeighborsRegressor

import numpy as np
import pandas as pd

data = {
    'student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'], 
    'hours_studied': [1,2,3,4,5], 
    'exam_score': [50, 60, 70, 80, 90]
}
df = pd.DataFrame(data)
X = df[['hours_studied']]
y = df['exam_score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
regressor = KNeighborsRegressor(n_neighbors=3)
regressor.fit(X_train, y_train)
hours = input("Enter number of hours studied: ")
predicted_score = regressor.predict([[float(hours)]])   
print(f"The predicted exam score is: {predicted_score[0]}")
```
